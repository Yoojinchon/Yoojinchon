## Let's figure out the Rating by Genre for what's prefer to Audience.

```python
# Average Rating by Genre

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("movies.csv")

genre_avg_rating = df.groupby('genre')['score'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
genre_avg_rating.plot(kind='bar', color='skyblue')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average Rating by Genre')
plt.xticks(rotation=45)
plt.show()
```

## Let's look at the trend in the film industry.

```python
# Total Profit by Year

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("movies.csv")

# Make a profit column (profit = gross - budget)
df['profit'] = df['gross'] - df['budget']

yearly_profit_sum = df.groupby('year')['profit'].sum()

plt.figure(figsize=(12, 6))
plt.plot(yearly_profit_sum.index, yearly_profit_sum.values, marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.title('Total Profit by Year (1980-2020)')
plt.grid(True)
plt.show()
```

## The most popular movie and its genre every year

```python


import pandas as pd

df = pd.read_csv("movies.csv")

top_gross_movies_by_year = df.groupby('year').apply(lambda x: x.loc[x['gross'].idxmax()])

top_gross_movies_by_year[['year', 'name', 'genre', 'gross']]
```

```python
# Group into 'year' and 'genre' and find genre with the highest gross for each year

import pandas as pd

df = pd.read_csv("movies.csv")

yearly_max_gross_genre = df.groupby('year')['gross'].idxmax()
top_grossing_genres_by_year = df.loc[yearly_max_gross_genre, ['year', 'genre']]

top_grossing_genres_by_year
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("movies.csv")

# Make a profit column (profit = gross - budget)
df['profit'] = df['gross'] - df['budget']

# Total Profit by Genre and Year
yearly_genre_profit_sum = df.groupby(['year', 'genre'])['profit'].sum().unstack(fill_value=0)

# visualization
plt.figure(figsize=(12, 8))
yearly_genre_profit_sum.plot(kind='bar', stacked=True, colormap='tab20c', figsize=(12, 8))
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.title('Total Profit by Genre and Year (1980-2020)')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y')
plt.show()
```

## Find out the popular genre according to the trend of the times

depends on the years, usually Action is the most high gross. Second, the animation is recording. Adventure is followed by a record. But successful of the movie is not the only for gross. Let's looking for the profit. I made the column profit = gross - budget

```python
import pandas as pd

df = pd.read_csv("movies.csv")

# add the 'profit' column (profit = gross - budget)
df['profit'] = df['gross'] - df['budget']

# make the group with year and profit
yearly_max_profit_genre = df.groupby('year')['profit'].idxmax()
top_profit_genres_by_year = df.loc[yearly_max_profit_genre, ['year', 'genre']]

print(top_profit_genres_by_year)
```

```python
#scatter plot

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.scatter(top_profit_genres_by_year['year'], top_profit_genres_by_year['genre'], s=50, c='blue', marker='o', edgecolor='black')
plt.xlabel('Year')
plt.ylabel('Genre')
plt.title('Top Profit Genre by Year')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("movies.csv")

corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
